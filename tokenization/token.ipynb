```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts

def merge(ids, pair, idx):
    first, second = pair
    new_ids = []
    i = 0
    while i < len(ids):
        if i < len(ids) - 1 and ids[i] == first and ids[i + 1] == second:
            new_ids.append(idx)
            i += 2
        else:
            new_ids.append(ids[i])
            i += 1

    return new_ids

vocab_size = 276
num_merges = vocab_size - 256
ids = list(tokens)

merges = {}

for i in range(num_merges):
    stats = get_stats(ids)
    if not stats:
        break
    best = max(stats, key=stats.get)
    print(f'Merge {i+1}: {best} -> {256 + i}, count: {stats[best]}')
    merges[best] = 256 + i
    ids = merge(ids, best, 256 + i)
```

```python
print(f'Compression: {len(tokens)} -> {len(ids)}')
```

```python
vocab = {idx: bytes([idx]) for idx in range(256)}
for (a, b), idx in merges.items():
    vocab[idx] = vocab[a] + vocab[b]

def decode(ids):
    tokens = []
    for idx in ids:
        tokens.append(vocab[idx])
    output = b''.join(tokens).decode('utf-8', errors='replace')
    return output

decoded_text = decode(ids)
print("Decoded successfully!")
print("Original length:", len(tokens))
print("Decoded length:", len(decoded_text.encode('utf-8')))
print("Match original:", decoded_text.encode('utf-8') == bytes(tokens))
```

```python
def encode(text):
    text_bytes = text.encode('utf-8')
    tokens = []
    for byte in text_bytes:
        tokens.append(byte)
    while len(tokens) > 1:
        stats = get_stats(tokens)
        pair = min(stats, key=lambda pair: merges.get(pair, float('inf')))
        if pair not in merges:
            break
        tokens = merge(tokens, pair, merges[pair])
    return tokens

print(encode("Hello, world!"))
print(decode(encode("Hello, world!")))
```
